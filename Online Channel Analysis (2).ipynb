```python
# Define a function to safely split the items list into separate columns for each item_id, price, and quantity
import json

def split_items_safe(data):
    # Initialize dictionaries to hold the new columns
    item_ids = {}
    prices = {}
    quantities = {}
    
    # Loop over each row in the data
    for index, row in data.iterrows():
        # Parse the items column
        items = row['items']
        try:
            if isinstance(items, str):  # If items are strings (e.g., from JSON structure), parse them
                items = json.loads(items.replace("'", "\""))
        except json.JSONDecodeError:
            # Skip rows with parsing errors
            continue

        # Loop over each item and add it to the dictionaries
        for i, item in enumerate(items, start=1):
            item_ids[f'Item{i}_id'] = item.get('item_id', None)
            prices[f'Item{i}_price'] = item.get('price', None)
            quantities[f'Item{i}_quantity'] = item.get('quantity', None)
        
        # Update the row with new columns
        for key, value in item_ids.items():
            data.loc[index, key] = value
        for key, value in prices.items():
            data.loc[index, key] = value
        for key, value in quantities.items():
            data.loc[index, key] = value

    return data

# Applying the transformation function to the online purchase data
online_transformed = split_items_safe(online_purchase_data.copy())

# Displaying the transformed online data to confirm the results
online_transformed.head()
```

```python
# Ensuring the metrics calculation is strictly for "purchase" events in the online data

# Filtering online data again for "purchase" events only to avoid any errors
online_purchase_data = online_data[online_data['event_name'] == 'purchase'].copy()

# Function to accurately calculate total sales value and items sold from 'items' column
def calculate_sales_metrics(data):
    total_sales_value = 0
    total_items_sold = 0
    for index, row in data.iterrows():
        try:
            items = json.loads(row['items'].replace("'", "\""))  # Parse items
            for item in items:
                price = item.get('price', 0)
                quantity = item.get('quantity', 1)  # Default quantity to 1 if missing
                total_sales_value += price * quantity
                total_items_sold += quantity
        except (json.JSONDecodeError, TypeError):
            continue  # Skip rows with errors in 'items' format
    return total_sales_value, total_items_sold

# Recalculating metrics based on purchases only in the online data
total_sales_value_online, total_items_sold_online = calculate_sales_metrics(online_purchase_data)

# Calculating average transaction value and average items per purchase
total_transactions_online = online_purchase_data['user_id'].nunique()
average_transaction_value_online = total_sales_value_online / total_transactions_online
average_items_per_purchase_online = total_items_sold_online / total_transactions_online

# Displaying the final results
total_sales_value_online, total_items_sold_online, average_transaction_value_online, average_items_per_purchase_online
```

```python
import pandas as pd
import json

# Load the online data and filter for 'purchase' events
online_data = pd.read_csv('/Users/farahfatima/Downloads/Marketing_AM_Assignment 2/web_events_item_brand.csv', low_memory=False)
online_purchase_data = online_data[online_data['event_name'] == 'purchase'].copy()

# Step 1: Parse the 'items' JSON data and extract individual item prices and quantities
def extract_prices_and_quantities(data):
    total_sales_value = 0
    total_items_sold = 0
    prices = []
    
    for _, row in data.iterrows():
        try:
            items = json.loads(row['items'].replace("'", "\""))  # Parse items as JSON
            for item in items:
                price = item.get('price', 0)
                quantity = item.get('quantity', 1)
                total_sales_value += price * quantity
                total_items_sold += quantity
                prices.extend([price] * quantity)  # Include each price multiple times based on quantity
        except (json.JSONDecodeError, TypeError):
            continue  # Skip rows with errors in 'items' format
    
    average_price = total_sales_value / total_items_sold if total_items_sold > 0 else 0
    return average_price, prices

# Calculate the average price and extract a list of all prices
average_price, all_prices = extract_prices_and_quantities(online_purchase_data)

# Step 2: Define price buckets for sensitivity analysis
price_bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1500, 2000, 3000, 4000]
price_bucket_series = pd.cut(all_prices, bins=price_bins)
price_bucket_df = pd.DataFrame({'price': all_prices, 'price_bucket': price_bucket_series})

# Step 3: Calculate total revenue for each price bucket
price_bucket_analysis = price_bucket_df.groupby('price_bucket').agg(total_revenue=('price', 'sum')).reset_index()

# Step 4: Calculate the average revenue across all price buckets
average_revenue = price_bucket_analysis['total_revenue'].mean()

# Step 5: Identify buckets with higher-than-average revenue
optimal_price_buckets = price_bucket_analysis[price_bucket_analysis['total_revenue'] > average_revenue]

# Display the results
print(f"Average Price: ${average_price:.2f}")
print("Optimal Price Points (Buckets with Higher Revenue):")
print(optimal_price_buckets[['price_bucket', 'total_revenue']])


```

```python
import pandas as pd
import json

# Load the online data and filter for 'purchase' events
online_data = pd.read_csv('/Users/farahfatima/Downloads/Marketing_AM_Assignment 2/web_events_item_brand.csv')
online_purchase_data = online_data[online_data['event_name'] == 'purchase'].copy()

# Function to safely parse items, calculate revenue, and quantity sold in each price bucket
def calculate_metrics(data):
    total_sales_value = 0
    total_items_sold = 0
    prices = []
    quantities = []

    for index, row in data.iterrows():
        try:
            items = json.loads(row['items'].replace("'", "\""))
            for item in items:
                price = item.get('price', 0)
                quantity = item.get('quantity', 1)
                total_sales_value += price * quantity
                total_items_sold += quantity
                prices.append(price)
                quantities.append(quantity)
        except json.JSONDecodeError:
            continue

    return total_sales_value, total_items_sold, prices, quantities

# Calculate total sales, items sold, and gather all prices and quantities
total_sales_value, total_items_sold, prices, quantities = calculate_metrics(online_purchase_data)

# Define price buckets for analysis and add total quantity for each bucket
price_bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1500, 2000, 3000, 4000]
online_purchase_data['price_bucket'] = pd.cut(pd.Series(prices), bins=price_bins)

# Create a DataFrame to calculate revenue and quantity per price bucket
bucket_data = pd.DataFrame({
    'price': prices,
    'quantity': quantities,
    'price_bucket': pd.cut(pd.Series(prices), bins=price_bins)
})

# Aggregate revenue and quantity for each price bucket
price_bucket_analysis = bucket_data.groupby('price_bucket').agg(
    total_revenue=('price', lambda x: (x * bucket_data.loc[x.index, 'quantity']).sum()),
    total_quantity=('quantity', 'sum')
).reset_index()

# Filter for top revenue-generating buckets
average_revenue = price_bucket_analysis['total_revenue'].mean()
top_buckets = price_bucket_analysis[price_bucket_analysis['total_revenue'] > average_revenue]

# Display results
print("Top Revenue-Generating Price Buckets with Quantities:")
print(top_buckets[['price_bucket', 'total_revenue', 'total_quantity']])
```

```python
import pandas as pd
import json

# Load the data and filter for 'purchase' events
online_data = pd.read_csv('/Users/farahfatima/Downloads/Marketing_AM_Assignment 2/web_events_item_brand.csv')
online_purchase_data = online_data[online_data['event_name'] == 'purchase'].copy()

# Define a function to parse items and calculate sales per brand
def calculate_brand_metrics(data):
    brand_sales = {}
    brand_quantity = {}

    for index, row in data.iterrows():
        try:
            items = json.loads(row['items'].replace("'", "\""))
            for item in items:
                brand = item.get('item_brand', 'Others')
                price = item.get('price', 0)
                quantity = item.get('quantity', 1)

                # Update sales and quantity for each brand
                if brand in brand_sales:
                    brand_sales[brand] += price * quantity
                    brand_quantity[brand] += quantity
                else:
                    brand_sales[brand] = price * quantity
                    brand_quantity[brand] = quantity
        except json.JSONDecodeError:
            continue

    # Convert to DataFrame
    brand_df = pd.DataFrame({
        'Brand': list(brand_sales.keys()),
        'Total_Sales_Value': list(brand_sales.values()),
        'Items_Sold': list(brand_quantity.values())
    })

    # Calculate Market Share
    total_sales = brand_df['Total_Sales_Value'].sum()
    brand_df['Market_Share'] = (brand_df['Total_Sales_Value'] / total_sales) * 100

    # Sort by Total Sales Value
    brand_df = brand_df.sort_values(by='Total_Sales_Value', ascending=False).reset_index(drop=True)
    
    return brand_df

# Run the analysis
brand_analysis_df = calculate_brand_metrics(online_purchase_data)

# Display top 5 brands and "Others" category
top_5_brands = brand_analysis_df.head(5)
others_market_share = brand_analysis_df[brand_analysis_df['Brand'] == 'Others']['Market_Share'].values[0]
total_market_share_top_5 = top_5_brands['Market_Share'].sum()

# Display results
print("Top 5 Brands Analysis:")
print(top_5_brands)
print(f"Top 5 brands control {total_market_share_top_5:.2f}% of online sales.")
print(f"'Others' category has a market share of {others_market_share:.2f}%, suggesting potential for brand expansion.")
```
